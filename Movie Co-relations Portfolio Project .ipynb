```python
#Importing Packages required
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configurations of the plots we create.

#Reading the data 

df = pd.read_csv(r"C:\Users\Neeraj\Downloads\movies.csv\movies.csv")
```

```python
#Glancing the data
df.head()
```

```python
df = df.dropna() #Dropping all the rows that contain a null value

#Checking for missing values in the dataset
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {:.2f}%'.format(col , pct_missing))
```

```python
df.dtypes #Checking the datatypes of the columns of the dataset
```

```python
#type casting
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df['correct_year'] = df['released'].astype(str).str.split(' ')
df['correct_year_1'] = df['correct_year'].str[2]
```

```python
#removing unwanted columns
df.drop('correct_year', axis = 1,inplace = True)
df
```

```python
#Sorting values 
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
#Building Co-relations between the colums of the dataset
#Building a scatter plot bugdet vs gross
plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget v/s Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
#Plotting using seaborn 
sns.regplot(x='budget', y= 'gross', data = df, scatter_kws = {"color" : "black"}, line_kws = {"color":"blue"}).set(title = "Budget v/s Gross Earnings")
            
# df.head()
```

```python
#Looking at the Co-relation
df.corr(method = 'pearson') #Only works on numerical fields/columns
#There are 3 types of corelations -> pearson, kendall, spearman and all give different results, By default the method is pearson
```

```python
correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix')
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
```

```python
df_numerized = df.copy()
for col_name in df_numerized:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs > 0.5)]
high_corr
```
